```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
netflix_df = pd.read_csv('netflix_data.csv')
```

```python
netflix_subset = netflix_df[netflix_df['type'] == 'Movie']
```

```python
netflix_subset
```

```python
netflix_movies = netflix_subset[['title', 'country', 'genre', 'release_year', 'duration']]
```

```python
netflix_movies
```

```python
netflix_movies_90s = netflix_movies[(netflix_movies['release_year'] >= 1990) & (netflix_movies['release_year'] <= 1999)]
```

```python
netflix_movies_90s
```

```python
short_movies = netflix_movies[netflix_movies['duration'] < 60]

short_movies.head()
```

#### the most frequent movie duration

```python
most_frequent_duration = netflix_movies_90s['duration'].mode()[0]
print(f"Most frequent movie duration in the 1990s: {most_frequent_duration} minutes")
```

```python
# Save the result as an integer variable called 'duration'

duration = int(most_frequent_duration)
```

#### Visualizing durations of movies in 90s

```python
plt.figure(figsize=(10, 6))
plt.hist(netflix_movies_90s['duration'], bins=range(0, 301, 10), edgecolor='black')
plt.title('Distribution of Movie Durations in the 1990s')
plt.xlabel('Duration (minutes)')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

#### Count the number of short action movies from the 1990s

```python
short_action_movies = netflix_movies_90s[(netflix_movies_90s['genre'].str.contains('Action')) & (netflix_movies_90s['duration'] < 90)]
short_action_count = short_action_movies.shape[0]
print(f"Number of short action movies (less than 90 minutes) from the 1990s: {short_action_count}")
```

#### Visualization for durations by genre and year

```python
colors = []

for index, row in netflix_movies_90s.iterrows():
    if 'Children' in row['genre']:
        colors.append('green')
    elif 'Documentaries' in row['genre']:
        colors.append('green')
    elif 'Stand-Up' in row['genre']:
        colors.append('blue')
    else:
        colors.append('red')

fig, ax = plt.subplots(figsize=(12, 8))
scatter = ax.scatter(netflix_movies_90s['release_year'], netflix_movies_90s['duration'], c=colors, alpha=0.5)
ax.set_title('Movie Duration by Year of Release (1990s)')
ax.set_xlabel('Release Year')
ax.set_ylabel('Duration (minutes)')
```

#### Creating a legend

```python
legend_labels = {
    'green': 'Children/Documentaries',
    'blue': 'Stand-Up',
    'red': 'Other Genres'
}
handles = [plt.Line2D([0], [0], marker='o', color='w', markerfacecolor=color, markersize=10, label=label) for color, label in legend_labels.items()]
ax.legend(handles=handles, loc='best')

plt.grid(True)
plt.show()
```
